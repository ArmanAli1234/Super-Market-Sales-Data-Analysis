```python
# remove useless features
df.drop(columns=["Order ID","Order Date","Ship Date"],inplace=True)
```

```python
df.drop(columns=["Country","City","State","Product ID","Sub-Category","Product Name","Order month"],axis=1,inplace=True)
```

```python
df.drop(columns=["Order year"],axis=1,inplace=True)
```

```python
df['Ship Mode'].value_counts()
```

```python
# apply cyclic encoding on month column,becaue it repeat once again after again ,but model wouldn't understand this so we use it.
df['month_sin']=np.sin(2*np.pi*df['Order num']/12)
df['month_cos']=np.cos(2*np.pi*df['Order num']/12)
```

```python
df.columns=df.columns.str.strip()
```

```python
df.drop(["Profit","Order num"],axis=1,inplace=True)
```

```python
# hist plot 
plt.figure(figsize=(9,7))
df.hist()
plt.tight_layout()
```

```python
# Outlier Detection
def outlier_detection(col):
    print("-------------OUTLIER DETECTION--------------")
    print(f"Outlier Detection for {col}")
    q1=df[col].quantile(0.25)
    q3=df[col].quantile(0.75)
    iqr=q3-q1
    lower_bound=q1-iqr*1.5
    upper_bound=q3+iqr*1.5

    # print value
    print(f"Quantile 1:- {q1}")
    print(f"Quantile 3:- {q3}")
    print(f"IQR : {iqr}")
    print(f"Lower Bound: {lower_bound}")
    print(f"Upper Bound: {upper_bound}")
    print(f"skew: {df[col].skew()}")


    # visualization
    plt.figure(figsize=(7,8))
    plt.subplot(221)
    sns.distplot(df[col])
    plt.title(f"Distribution of {col}")

    plt.subplot(222)
    sns.boxplot(df[col])
    plt.title(f"Boxplot of {col}")

    plt.tight_layout()
    plt.show()

```

```python
# ignore warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
# split num and cat columns
df_num=df.select_dtypes(include=[np.number])
df_cat=df.select_dtypes(exclude=[np.number])

#outlier detection
for col in df_num.columns:
    outlier_detection(col)
```

##### Sales:
##### Sales data is right-skewed with a few high-value outliers, indicating bulk or premium transactions contributing significantly to revenue.

##### Quantity:
##### Quantity shows moderate skewness with most values between 10–14, reflecting consistent and stable purchasing behavior.

##### Discount:
##### Discount is right-skewed with high-value outliers (0.6–0.8), likely representing promotional or clearance strategies.

##### Profit:
##### Profit is highly skewed with both positive and negative extremes, indicating variability due to pricing and discounting strategies.

## Feature Engineering

```python
# distributed the data into two parts training and testing data
from sklearn.model_selection import train_test_split
x=df.drop('Sales',axis=1)
y=df['Sales']
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
X_train
```

# Feature Engineering

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import cross_val_score
from sklearn.pipeline import Pipeline
```

```python
cat_col=["Ship Mode","Segment","Region","Category"]
num_col=["Quantity","Discount","month_sin","month_cos"]
```

```python
preprocessor=ColumnTransformer([
    ('num_cols',StandardScaler(),num_col),
    ('cat_cols',OneHotEncoder(),cat_col)
])
```

```python
X_train.columns
```

```python
X_train=preprocessor.fit_transform(X_train)
```

```python
X_test=preprocessor.transform(X_test)
```

```python
from sklearn.metrics import r2_score
from sklearn.model_selection import cross_val_score

def model_selection(model):
    
    model.fit(X_train, y_train)
    
    y_pred = model.predict(X_test)
    
    print(f"R2 Score (Test) of {name}: {r2_score(y_test, y_pred)}")
    
    cv_scores = cross_val_score(model, X_train, y_train, cv=10, scoring="r2")
    
    print(f"CV Mean Score of {name}: {cv_scores.mean()}")
    print(f"CV Std Dev of {name}: {cv_scores.std()}")
```
